```python
import pandas as pd
from kucoin.client import Market
import datetime as date
import time
import pytz
```

```python
#get remote to kucoin api
```

```python
client = Market(url='https://api.kucoin.com')
```

```python
#Collect Market Data and create data frame
```

```python

#get start and end unix times from 'farback' agp
def getTime(farback):
    date=datetime.now()
    #convert current time to unix
    end = time.mktime(date.timetuple()) 
    start = int(end - farback)
    end = int(end)
    return start, end


```

```python


def getData(pair='BTC-USDT', farback=3600, c_stick='1min'):
    start, end = getTime(farback)
    #create data frame from info given
    df = pd.DataFrame(client.get_kline(symbol=pair, startAt=start, endAt=end, kline_type=c_stick))
    df = df.iloc[:,:6]
    df.columns = ['Time', 'Open', 'High', 'Low', 'Close', 'Volume']
    #convert unix time to utc time
    df['Time'] = pd.to_datetime(df['Time'], unit='s', utc=True)
    df = df.set_index('Time')
    #convert utc time to est time
    df.index = df.index.tz_convert('America/New_York')
    df = df.astype(float)    
    print(df)

    

getData()
```
